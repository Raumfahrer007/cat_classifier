# Baseline Model Development
This notebook establishes our performance benchmark by implementing a custom Convolutional Neural Network from scratch. Before leveraging advanced techniques like transfer learning, we first build a baseline model to understand the fundamental challenges of our classification task and set a reference point for measuring improvement.

The baseline serves multiple critical purposes: it validates our data pipeline, reveals how a simple model handles class imbalance, identifies potential learning shortcuts, and provides insights into the problem's inherent difficulty. By starting with a basic architecture, we can better appreciate the value of more sophisticated approaches while ensuring our fundamental implementation is sound.

We'll implement a straightforward CNN architecture, apply the class weights calculated during our exploration, and establish key performance metrics that will guide our subsequent model iterations.

# 1. Imports

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

import numpy as np
import os
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import importlib
from dataclasses import dataclass

print("=== IMPORTS FINISHED ===")
```

# 2. Config

```python
@dataclass
class Config:
    IMG_HEIGHT: int = 224
    IMG_WIDTH: int = 224
    IMG_CHANNELS: int = 3
    
    # Data settings
    BATCH_SIZE: int = 32
    VALIDATION_SPLIT: float = 0.15
    TEST_SPLIT: float = 0.15
    
    # Training settings
    LEARNING_RATE: float = 1e-4
    EPOCHS: int = 10
    
    # Paths
    DATA_PATH: str = "../data/raw"
    PROCESSED_PATH: str = "../data/processed"
```

# 3. Data Preparation

## 3.1 Data Augmentation Configuration
We begin by defining our data preprocessing pipeline. The training generator includes aggressive augmentation to address the class imbalance and environmental biases identified in our exploration, while the validation and test generators perform only basic normalization to ensure fair evaluation.

```python
def create_data_generators(
    target_size=(Config.IMG_HEIGHT, Config.IMG_WIDTH),
    batch_size=Config.BATCH_SIZE
):
    """
    Create data generators with handling for class imbalance
    """
    # Training generator with heavy augmentation for minority classes
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=25,
        width_shift_range=0.3,
        height_shift_range=0.3,
        horizontal_flip=True,
        vertical_flip=True,
        zoom_range=0.3,
        brightness_range=[0.7, 1.3],
        shear_range=0.2,
        fill_mode='nearest'
    )
    
    val_datagen = ImageDataGenerator(rescale=1./255)

    test_datagen = ImageDataGenerator(rescale=1./255)
    
    return train_datagen, val_datagen, test_datagen
```

## 3.2. Creating Data Generators
With our augmentation strategy defined, we instantiate the data generators for training, validation, and testing. These generators will handle image resizing, batch processing, and real-time augmentation during model training.

```python
train_datagen, val_datagen, test_datagen = create_data_generators()

# Create data generators
baseline_train_generator = train_datagen.flow_from_directory(
    f'{Config.PROCESSED_PATH}/train',
    target_size=(Config.IMG_HEIGHT, Config.IMG_WIDTH),
    batch_size=Config.BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

baseline_val_generator = val_datagen.flow_from_directory(
    f'{Config.PROCESSED_PATH}/val', 
    target_size=(Config.IMG_HEIGHT, Config.IMG_WIDTH),
    batch_size=Config.BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

baseline_test_generator = test_datagen.flow_from_directory(
    f'{Config.PROCESSED_PATH}/test',
    target_size=(Config.IMG_HEIGHT, Config.IMG_WIDTH), 
    batch_size=Config.BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

baseline_classes = baseline_train_generator.class_indices

print(f"Classes: {baseline_classes}")
print(f"Training samples: {train_generator.samples}")
print(f"Validation samples: {val_generator.samples}")
print(f"Test samples: {test_generator.samples}")
```

## 3.3 Class Weights

To counter the class imbalance identified in our data exploration, we implement class weighting—a technique that increases the penalty for misclassifying minority class examples. This approach ensures the model pays appropriate attention to all classes during training, rather than being dominated by the majority class.

The weights are calculated using inverse frequency scaling, where each class's weight is proportional to its underrepresentation in the dataset.

```python
def calculate_class_weights(
    classes_dict,
    data_path=Config.DATA_PATH
):
    """
    Calculate class weights to handle imbalance during training
    Uses inverse frequency weighting
    """
    class_counts = {}
    
    # Count images for each class using the actual folder names
    for class_name, class_idx in classes_dict.items():
        class_path = os.path.join(data_path, class_name)
        image_files = [f for f in os.listdir(class_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
        class_counts[class_idx] = len(image_files)  # Use index as key for Keras
    
    total_samples = sum(class_counts.values())
    num_classes = len(class_counts)
    
    class_weights = {}
    for class_idx, count in class_counts.items():
        # Inverse frequency weighting
        class_weights[class_idx] = total_samples / (num_classes * count)
    
    return class_weights

# Calculate and display weights
baseline_weights = calculate_class_weights(baseline_classes)

print("Class weights for handling imbalance:")
for class_name, class_idx in baseline_classes.items():
    print(f"  {class_name}({class_idx}): {baseline_weights[class_idx]:.3f}")
```

Class weights calculated:

- both (0): 2.14× weight - Strong emphasis on our most underrepresented class

- karamela (1): 0.95× weight - Near-balanced representation

- lacta (2): 0.68× weight - Reduced emphasis on the majority class

These weights will be passed to our training process, effectively rebalancing the learning signal and encouraging the model to develop robust features for all three classes despite their uneven distribution in our dataset.

# 4. Baseline Model

With our data pipeline established and class imbalance addressed, we now construct a simple Convolutional Neural Network from scratch. This baseline model serves as our performance benchmark — a reference point against which we'll measure all future improvements.

The architecture follows a classic CNN pattern: convolutional layers for feature extraction followed by dense layers for classification. We start with a modest capacity to establish a meaningful baseline before advancing to more complex architectures.

This approach allows us to validate our data pipeline, understand the problem's inherent difficulty, and appreciate the value of more sophisticated techniques we'll implement later.

# 4.1 Building the Baseline CNN Architecture

```python
def create_baseline_model(
    num_classes,
    input_shape=(Config.IMG_HEIGHT, Config.IMG_WIDTH, Config.IMG_CHANNELS)
):
    """
    Create a simple CNN baseline model
    """
    model = models.Sequential([
        layers.Input(shape=input_shape),
         # First conv block: learns basic features (edges, textures)
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
         # Second conv block: learns more complex patterns
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        # Third conv block: learns high-level features
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        # Transition to classification
        layers.Flatten(), # Converts 3D features to 1D vector

        # Dense layers for decision making
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.5), # Prevents overfitting by randomly disabling neurons

        # Output layer: probability distribution over 3 classes
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

def compile_model(model, learning_rate=Config.LEARNING_RATE):
    """
    Compile model with appropriate settings
    """
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

print("=== BASELINE MODEL ===")
baseline_model = create_baseline_model(len(baseline_classes))
baseline_model = compile_model(baseline_model)

# Display model architecture
baseline_model.summary()
```

Our baseline CNN has been successfully constructed with 2.8 million parameters. The architecture reveals several important characteristics that will influence our training approach:

Key Architectural Insights:
Spatial Progression: The model systematically reduces image dimensions from 224×224 to 26×26 while increasing feature depth from 32 to 64 filters. This hierarchical processing allows the network to learn increasingly complex patterns—from basic edges to sophisticated cat features.

Parameter Distribution:

Convolutional layers: 56,320 parameters (2%) - Feature learning capacity

Dense layers: 2,769,155 parameters (98%) - Decision-making power

This distribution highlights that the majority of learning occurs in the final classification layers, which aligns with our expectation that distinguishing between Lacta and Karamela requires sophisticated feature combination.

Potential Challenge Identified: The Flatten layer produces 43,264 features feeding into the dense network. This high-dimensional transition could lead to overfitting, but our 50% dropout layer provides crucial regularization to mitigate this risk.

Training Implications:
The substantial parameter count suggests we have sufficient model capacity to learn complex cat distinctions

The progressive downsampling ensures computational efficiency while preserving important features

The architecture provides a meaningful benchmark against which we can measure transfer learning improvements

This baseline establishes our foundation—a capable but computationally intensive network that will help us quantify the value of more advanced approaches.

## 4.2 Baseline Model Training

With our architecture compiled and data pipeline prepared, we now initiate the training process. This phase represents the core of machine learning—where our model will iteratively learn to distinguish between Lacta, Karamela, and their co-appearances through exposure to our training data.

The training configuration incorporates several strategic elements we've previously established: class weights to address imbalance, validation monitoring to prevent overfitting, and a controlled number of epochs to ensure thorough learning without excessive computation.

During this process, the model will process batches of augmented images, make predictions, calculate errors using categorical crossentropy, and systematically adjust its 2.8 million parameters via the Adam optimizer. The validation set provides real-time performance feedback, allowing us to monitor generalization beyond the training data.

This training session establishes our crucial performance benchmark—a reference point that will help us quantify the improvements achieved through more advanced techniques in subsequent experiments.

```python
print("=== TRAINING BASELINE ===")
baseline_history = baseline_model.fit(
    baseline_train_generator,
    epochs=Config.EPOCHS,
    validation_data=baseline_val_generator,
    class_weight=baseline_weights,
    verbose=1
)
```

# 5. Baseline Model Performance Evaluation

With training completed, we now analyze our model's learning behavior and predictive capabilities. This evaluation phase serves multiple critical purposes: diagnosing learning patterns, quantifying performance across all classes, and establishing a comprehensive benchmark for future model comparisons.

The visualization of training curves will reveal how effectively our model learned, whether it overfit to the training data, and how well it generalizes to unseen examples. Subsequent analysis of per-class metrics will show how effectively our class weighting strategy addressed the imbalance, particularly for the challenging "both" category.

This comprehensive assessment transforms raw training logs into actionable insights, guiding our approach to more sophisticated architectures in subsequent experiments.

## 5.1 Training History Visualization
We begin by visualizing the learning curves to understand our model's training dynamics. The accuracy plot reveals how well the model learns from the training data and generalizes to validation examples, while the loss plot shows the optimization progress and potential overfitting patterns. These curves provide immediate visual feedback on whether our training strategy is effective.

```python
plt.figure(figsize=(15, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(baseline_history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(baseline_history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Baseline Model - Accuracy Over Time')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(baseline_history.history['loss'], label='Training Loss', marker='o')
plt.plot(baseline_history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Baseline Model - Loss Over Time')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print final metrics
final_train_acc = baseline_history.history['accuracy'][-1]
final_val_acc = baseline_history.history['val_accuracy'][-1]
final_train_loss = baseline_history.history['loss'][-1]
final_val_loss = baseline_history.history['val_loss'][-1]

print("=== FINAL TRAINING METRICS ===")
print(f"Training Accuracy: {final_train_acc:.3f}")
print(f"Validation Accuracy: {final_val_acc:.3f}")
print(f"Training Loss: {final_train_loss:.3f}")
print(f"Validation Loss: {final_val_loss:.3f}")
```

The learning curves reveal a baseline model that achieved modest but meaningful performance. With final training accuracy of 46.6% and validation accuracy of 48.2%, the model demonstrates it can learn to distinguish between our cats better than random chance (33.3%), though substantial improvement potential remains.

The close alignment between training and validation metrics (48.2% vs 46.6% accuracy, 0.993 vs 1.027 loss) indicates good generalization without significant overfitting—a positive sign given our limited dataset. The consistent downward trend in both training and validation loss suggests the model was still learning when training concluded, potentially benefiting from additional epochs.

## 5.2 Detailed Performance Analysis
Moving beyond overall metrics, we now examine per-class performance through a comprehensive classification report and confusion matrices. This granular analysis reveals how well our class weighting strategy addressed the imbalance and identifies specific confusion patterns between Lacta, Karamela, and "both" categories. The dual-format confusion matrix (counts and percentages) provides both absolute and relative performance perspectives.

```python
## Model Predictions and Detailed Metrics

baseline_class_names = baseline_classes.keys()

# Reset and get predictions from validation set
val_generator.reset()
y_true = val_generator.classes
y_pred = baseline_model.predict(val_generator)
y_pred_classes = np.argmax(y_pred, axis=1)

# Classification report with per-class metrics
print("=== DETAILED CLASSIFICATION REPORT ===")
print(classification_report(y_true, y_pred_classes, 
                          target_names=baseline_class_names, 
                          digits=3))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)
cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

plt.figure(figsize=(12, 5))

# Confusion matrix (counts)
plt.subplot(1, 2, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=baseline_class_names, yticklabels=baseline_class_names)
plt.title('Confusion Matrix (Counts)\nBaseline Model')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

# Confusion matrix (percentages)
plt.subplot(1, 2, 2)
sns.heatmap(cm_percent, annot=True, fmt='.1f', cmap='Blues',
            xticklabels=baseline_class_names, yticklabels=baseline_class_names)
plt.title('Confusion Matrix (%)\nBaseline Model')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

plt.tight_layout()
plt.show()
```

The classification report reveals critical patterns in how our model handles the class imbalance:

Striking Recall-Precision Tradeoff: The "both" class shows extreme recall (84.6%) but very low precision (21.6%), indicating the model frequently misclassifies individual cats as "both" while correctly identifying most actual "both" images. This suggests our class weighting may be overly aggressive for the minority class.

Majority Class Dominance: Lacta and Karamela show strong precision (89.5% and 92.3% respectively) but modest recall (41.5% and 41.4%), revealing the model is conservative—when it predicts these classes, it's usually correct, but it misses many true instances.

Macro vs Weighted Averages: The substantial gap between macro (49.4%) and weighted (53.3%) F1-scores confirms the imbalance's impact, with the model struggling most on the underrepresented "both" class despite our weighting strategy.

## 5.3 Model Persistence and Performance Documentation
To ensure reproducibility and enable future comparisons, we save the trained baseline model and compile a comprehensive performance summary. This documentation captures our benchmark results that will serve as a reference point for evaluating more advanced architectures. The saved model also allows for retrospective analysis and demonstration of our baseline capabilities.

```python
## Model Persistence and Performance Summary

# Save the baseline model for future reference
baseline_model.save('../models/baseline_model.keras')
print("Baseline model saved to '../models/baseline_model.keras'")

# Create performance summary
baseline_performance = {
    'final_train_accuracy': final_train_acc,
    'final_val_accuracy': final_val_acc,
    'final_train_loss': final_train_loss,
    'final_val_loss': final_val_loss,
    'per_class_metrics': classification_report(y_true, y_pred_classes, 
                                              target_names=baseline_classes.keys(), 
                                              output_dict=True)
}

print("\n🎯 BASELINE PERFORMANCE SUMMARY")
print(f"Final Validation Accuracy: {baseline_performance['final_val_accuracy']:.3f}")
print(f"Final Validation Loss: {baseline_performance['final_val_loss']:.3f}")
```

The baseline model has been successfully persisted, capturing our initial benchmark of 48.2% validation accuracy. This establishes a crucial reference point for evaluating future architectural improvements. The comprehensive performance documentation ensures we can make data-driven decisions when advancing to transfer learning approaches, with particular attention needed for rebalancing our class weighting strategy given the observed recall-precision imbalance.

## 6. Test Set Evaluation: Final Baseline Assessment
Having completed our development-phase analysis using the validation set, we now conduct the crucial final evaluation on our held-out test set. This represents the ultimate assessment of our baseline model's performance on completely unseen data, providing an unbiased estimate of how the model would perform in real-world scenarios.

The test set has remained untouched throughout all model development and analysis phases, ensuring no information leakage could influence our architectural decisions. This rigorous approach maintains the integrity of our evaluation pipeline and establishes a trustworthy benchmark for comparing future model iterations.

This final assessment will reveal whether our baseline model's performance generalizes consistently from validation to testing, completing the comprehensive evaluation of our initial approach before we advance to more sophisticated transfer learning techniques.

```python
print("=== FINAL BASELINE TEST EVALUATION ===")

baseline_test_generator.reset()
baseline_test_loss, baseline_test_accuracy = baseline_model.evaluate(baseline_test_generator)
print(f"Test Accuracy: {baseline_test_accuracy:.3f}")
print(f"Test Loss: {baseline_test_loss:.3f}")

# Compare with validation performance
print(f"Validation Accuracy: {baseline_performance['final_val_accuracy']:.3f}")
print(f"Performance Gap: {baseline_test_accuracy - baseline_performance['final_val_accuracy']:.3f}")
```

The final baseline assessment reveals an encouraging result: our model achieves 54.2% accuracy on the test set, outperforming the validation accuracy by 6 percentage points. This positive performance gap suggests our model generalizes well to completely unseen data and may have been slightly conservative in its validation performance.

The improved test performance (54.2% vs 48.2% validation) indicates that:

Our data splits were effectively stratified, with test set difficulty comparable to validation

The model learned meaningful features rather than memorizing validation patterns

The 50% dropout regularization successfully prevented overfitting

With a test loss of 0.959 showing consistent improvement over the final validation loss (0.993), we've established a solid baseline benchmark of 54.2% accuracy that any subsequent models must surpass to demonstrate meaningful improvement.

This successful baseline completion sets the stage for transfer learning, where we'll leverage pre-trained feature extractors to build upon this foundation and target the remaining 45.8% performance gap.
