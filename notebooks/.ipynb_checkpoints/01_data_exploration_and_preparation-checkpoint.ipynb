# 3. Series Analysis and Data Splitting Strategy
We begin by analyzing the natural grouping of images in our dataset to inform a rigorous splitting methodology. Images are organized into distinct series using the naming pattern 'class_name_seriesX (number)' for related images from the same photo session, while standalone images follow the pattern 'class_name (number)'.

**Series Analysis Approach**
- Extract and count distinct series identifiers across all classes
- Analyze series distribution and image density per series
- Identify the number of unique photographic scenarios available for splitting

This analysis reveals the fundamental structure of our dataset, showing how many truly independent visual scenarios we have for training, validation, and testing. Understanding this series distribution is crucial for designing splits that maintain both statistical robustness and real-world relevance.
Once we quantify the available series, we will implement series-aware splitting that keeps all images from the same photographic session within a single split, transforming our evaluation from measuring memorization of variations to testing genuine generalization to novel scenarios.

## 3.1 Data Distribution Analysis

```python
def analyze_series_distribution():

    analysis_results = {}
    
    for class_name in CLASS_NAMES:
        class_path = os.path.join(DATA_PATH, class_name)
        
        # Group images by series
        series_dict = defaultdict(list)
        standalone_images = []
        total_images = 0
        
        for filename in os.listdir(class_path):
            if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                continue
                
            total_images += 1
                
            # Check if this is part of a series
            if '_series' in filename:
                # Extract series identifier
                parts = filename.split('_series')
                if len(parts) > 1:
                    series_id = f"series{parts[1].split(' ')[0]}"  # Get the series number
                    series_dict[series_id].append(filename)
                else:
                    standalone_images.append(filename)
            else:
                # This is a standalone image
                standalone_images.append(filename)
        
        # Calculate metrics
        n_series = len(series_dict)
        n_standalone = len(standalone_images)
        avg_images_per_series = sum(len(images) for images in series_dict.values()) / n_series if n_series > 0 else 0
        
        # Store results
        analysis_results[class_name] = {
            'total_images': total_images,
            'total_series': n_series,
            'standalone_images': n_standalone,
            'avg_images_per_series': round(avg_images_per_series, 1),
            'series_size_distribution': [len(images) for images in series_dict.values()]
        }
        
        # Print summary
        print(f"\n{class_name.upper()} CLASS ANALYSIS:")
        print(f"  Total images: {total_images}")
        print(f"  Total series: {n_series}")
        print(f"  Standalone images: {n_standalone}")
        print(f"  Average images per series: {avg_images_per_series:.1f}")
        
        # Show series size distribution
        if series_dict:
            size_counts = defaultdict(int)
            for series in series_dict.values():
                size_counts[len(series)] += 1
            print(f"  Series size distribution: {dict(size_counts)}")
    
    return analysis_results

# Run the analysis
print("=== SERIES DISTRIBUTION ANALYSIS ===")
series_analysis = analyze_series_distribution()
```

### Distribution Analysis
The series analysis reveals distinct patterns across our three classes. The "both" class demonstrates the highest series density with only 16 series containing 85 images (average 4.8 images/series), indicating concentrated multi-cat photographic sessions. In contrast, "lacta" and "karamela" show more distributed patterns with 71 and 50 series respectively, featuring smaller average series sizes (3.1 and 3.4 images/series).

**Key Insights**
- "both" class scarcity: Limited to 16 unique photographic scenarios, making robust splitting challenging
- Series size diversity: All classes show varied series lengths from single images to large sessions (up to 12 images)
- Standalone images: Significant portions of individual images (49 lacta, 23 karamela, 8 both) that require careful grouping

## Next Step: Series-Aware Splitting
We will now implement series-aware partitioning that:
- Groups standalone images into batches to create meaningful evaluation units
- Ensures minimum series counts in each split, particularly for the data-scarce "both" class
- Maintains series integrity by keeping all images from the same photographic session together
- Preserves class balance while preventing data leakage across splits

This approach will transform our evaluation from measuring memorization of image variations to testing genuine generalization to novel photographic scenarios.

## 3.2 Data Splitting Strategy
Based on our series distribution analysis, we implement a rigorous splitting methodology designed to prevent data leakage and ensure meaningful evaluation of model generalization.

### Core Splitting Principles

**Series Integrity Preservation**
- All images from the same photographic series remain together in a single split
- Series are treated as atomic units during partitioning
- Prevents models from seeing variations of the same scenario across different splits

**Standalone Image Grouping**
- Individual standalone images are batched into groups of 3-4
- Creates meaningful evaluation units from isolated images
- Maintains the anti-leakage protection while ensuring practical split sizes

**Balanced Series Allocation**
- Minimum of 3 series guaranteed for each class in validation and test splits
- Proportional series distribution while protecting minority classes
- Random shuffling with fixed seed for reproducibility

### Implementation Approach
- Training: ~70% of series (minimum 12 series per class)
- Validation: ~15% of series (minimum 3 series per class)
- Test: ~15% of series (minimum 3 series per class)

This strategy transforms our dataset evaluation from measuring memorization of image variations to testing genuine generalization to completely novel photographic scenarios, providing a more realistic assessment of real-world performance.

```python
def create_series_aware_splits(test_size=0.15, val_size=0.15, random_seed=42):
    """
    Create train/val/test splits that keep image series together to prevent data leakage.
    
    Args:
        data_path: Path to raw data with class folders
        output_path: Where to save the processed splits
        test_size: Proportion of series for test set
        val_size: Proportion of series for validation set  
        random_seed: Random seed for reproducibility
    """
    random.seed(random_seed)
    
    # Create output directories
    for split in ['train', 'val', 'test']:
        for class_name in CLASS_NAMES:
            os.makedirs(os.path.join(OUTPUT_PATH, split, class_name), exist_ok=True)
    
    for class_name in CLASS_NAMES:
        class_path = os.path.join(DATA_PATH, class_name)
        
        # Group images by series
        series_dict = defaultdict(list)
        standalone_images = []
        
        for filename in os.listdir(class_path):
            if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                continue
                
            # Check if this is part of a series
            if '_series' in filename:
                # Extract series identifier (e.g., "series1" from "lacta_series1 (1).jpg")
                parts = filename.split('_series')
                if len(parts) > 1:
                    series_id = f"series{parts[1].split(' ')[0]}"  # Get the series number
                    series_dict[series_id].append(filename)
                else:
                    standalone_images.append(filename)
            else:
                # This is a standalone image - treat each as its own "series"
                standalone_images.append(filename)

        # Group standalone images into batches
        batch_size = 4
        for i in range(0, len(standalone_images), batch_size):
            batch = standalone_images[i:i + batch_size]
            series_dict[f"standalone_batch_{i//batch_size}"] = batch
        
        # Get all series and shuffle
        all_series = list(series_dict.keys())
        random.shuffle(all_series)
        
        # Calculate split sizes
        n_series = len(all_series)
        n_test = max(3, int(n_series * test_size))
        n_val = max(3, int(n_series * val_size))
        n_train = n_series - n_test - n_val
        
        # Split series into sets
        test_series = all_series[:n_test]
        val_series = all_series[n_test:n_test + n_val]
        train_series = all_series[n_test + n_val:]
        
        # Copy files to respective splits
        split_counts = {'train': 0, 'val': 0, 'test': 0}
                
        for series_id in val_series:
            for filename in series_dict[series_id]:
                src = os.path.join(class_path, filename)
                dst = os.path.join(OUTPUT_PATH, 'val', class_name, filename)
                shutil.copy2(src, dst)
                split_counts['val'] += 1
                
        for series_id in test_series:
            for filename in series_dict[series_id]:
                src = os.path.join(class_path, filename)
                dst = os.path.join(OUTPUT_PATH, 'test', class_name, filename)
                shutil.copy2(src, dst)
                split_counts['test'] += 1

        for series_id in train_series:
            for filename in series_dict[series_id]:
                src = os.path.join(class_path, filename)
                dst = os.path.join(OUTPUT_PATH, 'train', class_name, filename)
                shutil.copy2(src, dst)
                split_counts['train'] += 1

        class_total = split_counts['train'] + split_counts['val'] + split_counts['test']
        train_pct = (split_counts['train'] / class_total) * 100
        val_pct = (split_counts['val'] / class_total) * 100  
        test_pct = (split_counts['test'] / class_total) * 100
        
        print(f"{class_name}:")
        print(f"  Series: {n_series} (train: {len(train_series)}, val: {len(val_series)}, test: {len(test_series)})")
        print(f"  Avg images per series: {sum(len(series_dict[s]) for s in all_series) / len(all_series):.1f}")
        print(f"  Images: {split_counts['train']} train ({train_pct:.1f}%), "
              f"{split_counts['val']} val ({val_pct:.1f}%), "
              f"{split_counts['test']} test ({test_pct:.1f}%)")

# Execute the splitting
print("Creating series-aware splits...")
create_series_aware_splits()
print("Splitting complete!")

# Verify the results
print("\n=== FINAL SPLIT SUMMARY ===")
total_all = total_counts['train'] + total_counts['val'] + total_counts['test']
train_percent = (total_counts['train'] / total_all) * 100
val_percent = (total_counts['val'] / total_all) * 100
test_percent = (total_counts['test'] / total_all) * 100

print(f"\nTotal: {total_counts['train']} train ({train_percent:.1f}%), "
      f"{total_counts['val']} val ({val_percent:.1f}%), "
      f"{total_counts['test']} test ({test_percent:.1f}%)")
```

### Split Analysis
The series-aware partitioning successfully created balanced splits while maintaining rigorous anti-leakage protection. The distribution closely approximates our target 70/15/15 ratio across all classes, with overall totals of 390 training (71.4%), 84 validation (15.4%), and 72 test (13.2%) images.

**Class-Specific Insights**
- "both" class demonstrates the highest series density (4.7 images/series) but lowest series count (18 total), resulting in the smallest absolute test set (9 images)
- "lacta" and "karamela" show more distributed patterns with 3.2-3.4 images/series, providing robust validation and test coverage
- All classes maintain minimum viable series counts (3+ in validation/test) for meaningful evaluation

This split ensures models are evaluated on truly novel photographic scenarios rather than variations of training images, providing a realistic assessment of generalization capability despite the inherent challenge of limited "both" class series.

# 4.2 Training Set Image Properties Analysis

Moving beyond class distribution, we now examine the technical characteristics of our training set images. Understanding image dimensions and properties is crucial for determining appropriate preprocessing steps and model architecture decisions that align with our actual training data.

We sample 20 images from each class within the training set to efficiently analyze size patterns while maintaining computational practicality. # 4.4 Training Set Color Distribution Analysis
Understanding color distribution across classes within our training set helps identify potential biases and ensures our model learns meaningful features rather than spurious color correlations. We analyze dominant colors and overall color characteristics from the training data to detect if certain classes are associated with specific color palettes due to shooting conditions or environments that our models will encounter during learning.

```python
def analyze_color_distribution(class_name, sample_size=20):
    """Analyze dominant colors for a given class"""
    class_path = os.path.join(TRAIN_DIR, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    dominant_colors = []
    for img_file in image_files[:sample_size]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                # Resize for efficiency and get dominant color
                img_small = img.resize((50, 50))
                img_array = np.array(img_small)
                # Calculate mean color (simplified approach)
                mean_color = np.mean(img_array, axis=(0, 1))
                dominant_colors.append(mean_color)
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
    
    return np.array(dominant_colors)

# Compare average RGB values across classes
rgb_means = {}
for class_name in CLASS_NAMES:
    colors = analyze_color_distribution(class_name)
    rgb_means[class_name] = np.mean(colors, axis=0)

# Bar plot comparison
plt.figure(figsize=(10, 6))
x_pos = np.arange(len(CLASS_NAMES))
width = 0.25

for i, (channel, color) in enumerate(zip(['Red', 'Green', 'Blue'], ['red', 'green', 'blue'])):
    means = [rgb_means[cls][i] for cls in CLASS_NAMES]
    plt.bar(x_pos + i*width, means, width, label=channel, color=color, alpha=0.7)

plt.xlabel('Class')
plt.ylabel('Mean Intensity')
plt.title('Average RGB Values by Class')
plt.xticks(x_pos + width, CLASS_NAMES)
plt.legend()
plt.show()

# Print statistical summary
print("Color Distribution Summary:")
for class_name in CLASS_NAMES:
    print(f"{class_name}: R={rgb_means[class_name][0]:.1f}, G={rgb_means[class_name][1]:.1f}, B={rgb_means[class_name][2]:.1f}")
```

Our color distribution analysis reveals systematic differences in image characteristics across classes within our training set:

**Class-Specific Color Signatures**
- **Lacta**: Balanced RGB distribution (113, 110, 101) with slight red dominance
- **Karamela**: Reduced blue channel intensity (112, 102, 90) creating warmer tones
- **"Both"**: More balanced but slightly muted color profile (105, 102, 99)

These color patterns present a potential learning shortcut - the model could mistakenly associate classification with color temperature and brightness variations rather than learning to recognize the cats' distinctive features and spatial relationships. The systematic differences in color profiles across classes might lead the model to rely on environmental lighting cues rather than actual cat detection.

To ensure robust feature learning, we will implement aggressive color augmentation in our preprocessing pipeline:
- **Color jittering**: Random adjustments to brightness, contrast, saturation, and hue
- **Normalization**: Standardize color distributions across all classes
- **Diverse augmentation**: Ensure the model encounters varied lighting conditions during training

This approach forces the model to focus on shape, texture, and spatial features - the actual distinguishing characteristics between Lacta, Karamela, and their co-appearances - rather than superficial color correlations present in the training data.

# 4.5 Training Set Brightness & Contrast Analysis

Examining brightness and contrast variations within our training set helps us understand lighting conditions and identify potential biases that could influence model learning. Significant differences in illumination between classes might lead the model to learn lighting patterns present in the training data rather than meaningful feline features and spatial relationships.

```python
def analyze_brightness_contrast(class_name, sample_size=20):
    """Analyze brightness and contrast for a given class"""
    class_path = os.path.join(TRAIN_DIR, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    brightness_values = []
    contrast_values = []
    
    for img_file in image_files[:sample_size]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                # Convert to grayscale for brightness/contrast analysis
                img_gray = img.convert('L')
                img_array = np.array(img_gray)
                
                # Calculate brightness (mean intensity)
                brightness = np.mean(img_array)
                brightness_values.append(brightness)
                
                # Calculate contrast (standard deviation)
                contrast = np.std(img_array)
                contrast_values.append(contrast)
                
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
    
    return brightness_values, contrast_values

# Analyze and compare across classes
plt.figure(figsize=(15, 8))

# Brightness analysis
plt.subplot(2, 2, 1)
brightness_data = []
for class_name in CLASS_NAMES:
    brightness, _ = analyze_brightness_contrast(class_name)
    brightness_data.append(brightness)
    plt.hist(brightness, alpha=0.7, label=class_name, bins=15)
plt.xlabel('Brightness (Mean Intensity)')
plt.ylabel('Frequency')
plt.title('Brightness Distribution by Class')
plt.legend()

# Contrast analysis
plt.subplot(2, 2, 2)
contrast_data = []
for class_name in CLASS_NAMES:
    _, contrast = analyze_brightness_contrast(class_name)
    contrast_data.append(contrast)
    plt.hist(contrast, alpha=0.7, label=class_name, bins=15)
plt.xlabel('Contrast (Standard Deviation)')
plt.ylabel('Frequency')
plt.title('Contrast Distribution by Class')
plt.legend()

# Box plot for brightness comparison
plt.subplot(2, 2, 3)
plt.boxplot(brightness_data, tick_labels=CLASS_NAMES)
plt.ylabel('Brightness')
plt.title('Brightness Distribution Comparison')

# Box plot for contrast comparison
plt.subplot(2, 2, 4)
plt.boxplot(contrast_data, tick_labels=CLASS_NAMES)
plt.ylabel('Contrast')
plt.title('Contrast Distribution Comparison')

plt.tight_layout()
plt.show()

# Print statistical summary
print("Brightness & Contrast Summary:")
for i, class_name in enumerate(CLASS_NAMES):
    print(f"{class_name}: Brightness ={np.mean(brightness_data[i]):.1f}, ={np.std(brightness_data[i]):.1f} | "
          f"Contrast ={np.mean(contrast_data[i]):.1f}, ={np.std(contrast_data[i]):.1f}")
```

Our illumination analysis reveals distinct patterns in lighting conditions across classes within our training set:

**Brightness & Contrast Patterns**
- **Brightness Distribution**: Lacta images are brightest (=109.7), followed by Karamela (=103.5) and "both" class (=102.2)
- **Contrast Levels**: All classes show similar contrast (59-64 range), with "both" class having slightly higher contrast (=64.3)
- **Variability Insights**: Karamela shows highest brightness variability (=21.6), while "both" class has the most consistent lighting (=13.5)

The systematic brightness differences present a potential learning shortcutthe model could learn to associate higher brightness with Lacta and lower brightness with multi-cat scenarios rather than recognizing actual feline features. The consistent contrast levels across classes is positive, indicating comparable image quality and edge definition without needing specialized contrast enhancement.

This finding reinforces our need for robust brightness augmentation in training:
- **Random brightness adjustments** to eliminate illumination-based shortcuts
- **Exposure variation** to ensure the model learns lighting-invariant features
- **Combined strategy** with color augmentation to force focus on structural characteristics

The training data's natural lighting variations provide a realistic foundation, while our augmentation strategy will ensure models learn robust, lighting-invariant representations of our feline subjects.

# 4.6 Training Set Background Complexity Analysis
Understanding background diversity within our training set helps us assess whether our model might learn to recognize environments rather than cats. If certain classes in the training data are predominantly associated with specific backgrounds, the model could develop spurious correlations based on environmental context rather than learning the actual feline features that distinguish each class.

```python
def analyze_background_complexity(class_name, sample_size=20):
    """Analyze background complexity using edge detection and texture analysis"""
    class_path = os.path.join(TRAIN_DIR, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    complexity_scores = []
    
    for img_file in image_files[:sample_size]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                # Resize for efficiency and convert to grayscale
                img_small = img.resize((200, 200)).convert('L')
                img_array = np.array(img_small)
                
                # Calculate edge density (proxy for background complexity)
                edges = ndimage.sobel(img_array)
                edge_density = np.mean(np.abs(edges))
                
                # Calculate texture variance
                texture_variance = np.var(img_array)
                
                # Combined complexity score
                complexity = edge_density * texture_variance
                complexity_scores.append(complexity)
                
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
    
    return complexity_scores

# Analyze background complexity across classes
plt.figure(figsize=(12, 5))

complexity_data = []
for class_name in CLASS_NAMES:
    complexities = analyze_background_complexity(class_name)
    complexity_data.append(complexities)

# Box plot comparison
plt.subplot(1, 2, 1)
plt.boxplot(complexity_data, tick_labels=CLASS_NAMES)
plt.ylabel('Background Complexity Score')
plt.title('Background Complexity by Class')

# Violin plot for distribution shape
plt.subplot(1, 2, 2)
violin_parts = plt.violinplot(complexity_data, showmedians=True)
plt.xticks([1, 2, 3], CLASS_NAMES)
plt.ylabel('Background Complexity Score')
plt.title('Complexity Distribution Shape')

# Color the violins
for pc, color in zip(violin_parts['bodies'], ['lightblue', 'lightcoral', 'lightgreen']):
    pc.set_facecolor(color)
    pc.set_alpha(0.7)

plt.tight_layout()
plt.show()

# Statistical summary
print("Background Complexity Summary:")
for i, class_name in enumerate(CLASS_NAMES):
    print(f"{class_name}: ={np.mean(complexity_data[i]):.2f}, ={np.std(complexity_data[i]):.2f}, "
          f"range=[{np.min(complexity_data[i]):.2f}, {np.max(complexity_data[i]):.2f}]")
```

Our background analysis reveals notable differences in environmental complexity across classes within our training set:

**Background Complexity Patterns**
- **Both**: Highest complexity (=494,245) with moderate variability, suggesting more detailed environments when both cats appear together
- **Karamela**: Moderate complexity (=396,765) with high variability, indicating diverse background settings
- **Lacta**: Lower complexity (=390,128) with substantial range, often photographed in less cluttered environments

The complexity differences present a potential learning shortcut - the model could associate highly complex backgrounds with the "both" class and simpler environments with individual cats, rather than learning the actual feline features and spatial relationships. The 27% complexity difference between "both" and Lacta backgrounds provides a meaningful but superficial signal the model might exploit.

This finding informs our data augmentation strategy for training:
- **Random cropping** to force focus on cat regions rather than full background context
- **Background augmentation** to ensure the model learns background-invariant features
- **Combined approach** with previous augmentations to create robust, environment-agnostic recognition

The substantial variability within each class is beneficial - it suggests enough background diversity in the training data to encourage the model to focus on the cats themselves rather than environmental cues present in the training samples.
