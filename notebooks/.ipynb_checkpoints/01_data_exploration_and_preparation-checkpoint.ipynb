We begin with 546 personally collected images featuring our two feline subjects, Lacta (black cat) and Karamela (orange cat), across three classification categories: individual appearances of each cat and images containing both cats together. Through systematic analysis of class distribution, image properties, and data quality, we lay the foundation for building a robust classifier capable of handling real-world image recognition challenges, including class imbalance and visual complexity.

## Imports

```python
import os
import shutil
import numpy as np
from PIL import Image
from scipy import ndimage
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

print("=== IMPORTS FINISHED ===")
```

# 2.Setup

```python
DATA_PATH = "../data/raw"
CLASS_NAMES = ['lacta', 'karamela', 'both']

print("=== SETUP FINISHED ===")
```

# 3. Class Distribution

We begin by quantifying our dataset's composition - counting how many images we have for each of our three classes: Lacta, Karamela, and Both cats together.

```python
class_counts = {}
for class_name in CLASS_NAMES:
    class_path = os.path.join(DATA_PATH, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    class_counts[class_name] = len(image_files)
    print(f"{class_name}: {len(image_files)} images")

# Plot distribution
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(class_counts.keys(), class_counts.values())
plt.title('Class Distribution')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.pie(class_counts.values(), labels=class_counts.keys(), autopct='%1.1f%%')
plt.title('Class Proportions')
plt.tight_layout()
plt.show()
```

Our initial analysis reveals a clear class imbalance across the 546-image dataset. Lacta represents the majority class with 269 images (49.3%), followed by Karamela with 192 images (35.2%), while the "both" category forms a distinct minority with only 85 examples (15.6%).

This distribution presents a classic machine learning challenge: while we have sufficient data for individual cat recognition, the limited examples of both cats together could hinder the model's ability to learn the complex spatial relationships and occlusions that occur when both felines appear in the same image.

Despite this imbalance, I've chosen to implement a three-class structure rather than separate binary classifiers. This approach allows the model to directly learn the nuanced visual patterns specific to scenarios where both cats appear together—information that cannot be easily inferred from individual cat recognition alone.

To address the imbalance, we will implement several strategic countermeasures: 
- class weighting during training to increase the cost of misclassifying minority examples
- targeted data augmentation to artificially expand our "both" class representation
- evaluation metrics that account for class imbalance rather than relying solely on overall accuracy.

Our data splits will maintain this proportional distribution to ensure realistic performance assessment.

# 4. Image Properties Analysis
Moving beyond class distribution, we now examine the technical characteristics of our image dataset. Understanding image dimensions and properties is crucial for determining appropriate preprocessing steps and model architecture decisions.

We sample 20 images from each class to efficiently analyze size patterns while maintaining computational practicality. This sampling approach provides representative insights without processing all 546 images at this exploratory stage.

```python
all_sizes = []
color_stats = {}

for class_name in CLASS_NAMES:
    class_path = os.path.join(DATA_PATH, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    # Check first 20 images per class to save time
    sizes = []
    for img_file in image_files[:20]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                sizes.append(img.size)  # (width, height)
        except Exception as e:
            print(f"Error reading {img_path}: {e}")
    
    if sizes:
        avg_width = np.mean([s[0] for s in sizes])
        avg_height = np.mean([s[1] for s in sizes])
        print(f"{class_name}: Avg size {avg_width:.0f}x{avg_height:.0f}")
        all_sizes.extend(sizes)

# Plot size distribution
if all_sizes:
    widths = [s[0] for s in all_sizes]
    heights = [s[1] for s in all_sizes]
    
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.scatter(widths, heights, alpha=0.6)
    plt.xlabel('Width')
    plt.ylabel('Height')
    plt.title('Image Dimensions')
    
    plt.subplot(1, 2, 2)
    plt.hist(widths, bins=20, alpha=0.7, label='Widths')
    plt.hist(heights, bins=20, alpha=0.7, label='Heights')
    plt.xlabel('Pixels')
    plt.ylabel('Frequency')
    plt.legend()
    plt.title('Size Distribution')
    plt.tight_layout()
    plt.show()
```

The consistency in image dimensions across all three classes is notable—all averaging around 4000×3000 pixels. This high-resolution uniformity simplifies our preprocessing pipeline, as we don't need to handle extreme size variations between classes. However, these dimensions are substantially larger than typical deep learning model inputs, indicating we'll need significant downsampling to balance computational efficiency with feature preservation.

The scatter plot and histogram visualizations confirm this size consistency while helping identify any outliers that might require special handling during preprocessing. This dimensional analysis directly informs our next step: selecting an appropriate input size for our neural network architecture.

# 5.Sample Images

We now examine sample images from each class to understand the visual characteristics and challenges our model will face. This qualitative assessment reveals the diversity in poses, lighting, backgrounds, and scales across our dataset.

```python
def display_samples(class_name, num_samples=5):
    class_path = os.path.join(DATA_PATH, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    plt.figure(figsize=(15, 3))
    plt.suptitle(f'Sample Images: {class_name}', fontsize=16)
    
    for i, img_file in enumerate(image_files[:num_samples]):
        img_path = os.path.join(class_path, img_file)
        try:
            img = Image.open(img_path)
            plt.subplot(1, num_samples, i+1)
            plt.imshow(img)
            plt.title(f"{img_file[:15]}...\n{img.size[0]}x{img.size[1]}")
            plt.axis('off')
        except Exception as e:
            print(f"Error displaying {img_path}: {e}")
    
    plt.tight_layout()
    plt.show()

for class_name in CLASS_NAMES:
    display_samples(class_name)
```

The "both" class presents additional complexity with spatial interactions between the cats, validating that this requires dedicated learning beyond individual cat recognition. # 6. Color Distribution Analysis
Understanding color distribution across classes helps identify potential biases and ensures our model learns meaningful features rather than spurious color correlations. We analyze dominant colors and overall color characteristics to detect if certain classes are associated with specific color palettes due to shooting conditions or environments.

```python
def analyze_color_distribution(class_name, sample_size=20):
    """Analyze dominant colors for a given class"""
    class_path = os.path.join(DATA_PATH, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    dominant_colors = []
    for img_file in image_files[:sample_size]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                # Resize for efficiency and get dominant color
                img_small = img.resize((50, 50))
                img_array = np.array(img_small)
                # Calculate mean color (simplified approach)
                mean_color = np.mean(img_array, axis=(0, 1))
                dominant_colors.append(mean_color)
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
    
    return np.array(dominant_colors)

# Compare average RGB values across classes
rgb_means = {}
for class_name in CLASS_NAMES:
    colors = analyze_color_distribution(class_name)
    rgb_means[class_name] = np.mean(colors, axis=0)

# Bar plot comparison
plt.figure(figsize=(10, 6))
x_pos = np.arange(len(CLASS_NAMES))
width = 0.25

for i, (channel, color) in enumerate(zip(['Red', 'Green', 'Blue'], ['red', 'green', 'blue'])):
    means = [rgb_means[cls][i] for cls in CLASS_NAMES]
    plt.bar(x_pos + i*width, means, width, label=channel, color=color, alpha=0.7)

plt.xlabel('Class')
plt.ylabel('Mean Intensity')
plt.title('Average RGB Values by Class')
plt.xticks(x_pos + width, CLASS_NAMES)
plt.legend()
plt.show()
```

Our color distribution analysis reveals systematic differences in image characteristics across classes:
- Overall Warm Lighting: All classes show red-dominant color profiles, indicating consistent warm lighting conditions

**Class-Specific Color Signatures**
- Lacta images show balanced RGB distribution (∼115, 110, 100)
- Karamela images exhibit reduced blue channel intensity (∼110, 105, 90)
- "Both" class displays significantly higher intensity across all channels (∼135, 130, 120)

These color patterns present a potential learning shortcut — the model could mistakenly associate classification with overall brightness and color temperature rather than learning to recognize the cats' distinctive features and spatial relationships. The "both" class's consistently higher brightness might lead the model to rely on illumination cues rather than actual cat detection.

To ensure robust feature learning, we will implement aggressive color augmentation in our preprocessing pipeline:
- Color jittering: Random adjustments to brightness, contrast, saturation, and hue
- Normalization: Standardize color distributions across all classes
- Diverse augmentation: Ensure the model encounters varied lighting conditions during training

This approach forces the model to focus on shape, texture, and spatial features — the actual distinguishing characteristics between Lacta, Karamela, and their co-appearances — rather than superficial color correlations.

# 7. Brightness & Contrast Analysis
Examining brightness and contrast variations helps us understand lighting conditions across our dataset and identify potential biases that could influence model performance. Significant differences in illumination between classes might lead the model to learn lighting patterns rather than meaningful feline features.

```python
def analyze_brightness_contrast(class_name, sample_size=20):
    """Analyze brightness and contrast for a given class"""
    class_path = os.path.join(DATA_PATH, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    brightness_values = []
    contrast_values = []
    
    for img_file in image_files[:sample_size]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                # Convert to grayscale for brightness/contrast analysis
                img_gray = img.convert('L')
                img_array = np.array(img_gray)
                
                # Calculate brightness (mean intensity)
                brightness = np.mean(img_array)
                brightness_values.append(brightness)
                
                # Calculate contrast (standard deviation)
                contrast = np.std(img_array)
                contrast_values.append(contrast)
                
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
    
    return brightness_values, contrast_values

# Analyze and compare across classes
plt.figure(figsize=(15, 8))

# Brightness analysis
plt.subplot(2, 2, 1)
brightness_data = []
for class_name in CLASS_NAMES:
    brightness, _ = analyze_brightness_contrast(class_name)
    brightness_data.append(brightness)
    plt.hist(brightness, alpha=0.7, label=class_name, bins=15)
plt.xlabel('Brightness (Mean Intensity)')
plt.ylabel('Frequency')
plt.title('Brightness Distribution by Class')
plt.legend()

# Contrast analysis
plt.subplot(2, 2, 2)
contrast_data = []
for class_name in CLASS_NAMES:
    _, contrast = analyze_brightness_contrast(class_name)
    contrast_data.append(contrast)
    plt.hist(contrast, alpha=0.7, label=class_name, bins=15)
plt.xlabel('Contrast (Standard Deviation)')
plt.ylabel('Frequency')
plt.title('Contrast Distribution by Class')
plt.legend()

# Box plot for brightness comparison
plt.subplot(2, 2, 3)
plt.boxplot(brightness_data, tick_labels=CLASS_NAMES)
plt.ylabel('Brightness')
plt.title('Brightness Distribution Comparison')

# Box plot for contrast comparison
plt.subplot(2, 2, 4)
plt.boxplot(contrast_data, tick_labels=CLASS_NAMES)
plt.ylabel('Contrast')
plt.title('Contrast Distribution Comparison')

plt.tight_layout()
plt.show()

# Print statistical summary
print("Brightness & Contrast Summary:")
for i, class_name in enumerate(CLASS_NAMES):
    print(f"{class_name}: Brightness μ={np.mean(brightness_data[i]):.1f}, σ={np.std(brightness_data[i]):.1f} | "
          f"Contrast μ={np.mean(contrast_data[i]):.1f}, σ={np.std(contrast_data[i]):.1f}")
```

Our illumination analysis reveals clear patterns in lighting conditions across classes:
- Brightness Gradient: "both" class images are significantly brighter (μ=125.9) than individual cat images, with Lacta (108.8) brighter than Karamela (102.7)
- Contrast Consistency: All classes show similar contrast levels (56-62 range), indicating comparable image quality and edge definition
- Variability Differences: Karamela images show the highest brightness variability (σ=24.7), suggesting diverse lighting conditions

The systematic brightness differences present a potential shortcut—the model could learn to associate higher brightness with the "both" class rather than recognizing the actual presence of two cats. The 23-point brightness gap between "both" and Karamela images is particularly concerning as it provides a strong but superficial signal.

This finding reinforces our need for robust brightness augmentation:
- Random brightness adjustments during training to eliminate illumination-based shortcuts
- Exposure variation to ensure the model learns invariant to lighting conditions
- Combined with our color augmentation, this creates a comprehensive preprocessing strategy that forces the model to focus on structural features rather than environmental lighting cues

The consistent contrast levels across classes is positive, indicating we don't need specialized contrast enhancement and can maintain natural image characteristics.

# 8. Background Complexity Analysis
Understanding background diversity helps us assess whether our model might learn to recognize environments rather than cats. If certain classes are predominantly associated with specific backgrounds, the model could develop spurious correlations that harm real-world performance.

```python
def analyze_background_complexity(class_name, sample_size=20):
    """Analyze background complexity using edge detection and texture analysis"""
    class_path = os.path.join(DATA_PATH, class_name)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    complexity_scores = []
    
    for img_file in image_files[:sample_size]:
        img_path = os.path.join(class_path, img_file)
        try:
            with Image.open(img_path) as img:
                # Resize for efficiency and convert to grayscale
                img_small = img.resize((200, 200)).convert('L')
                img_array = np.array(img_small)
                
                # Calculate edge density (proxy for background complexity)
                edges = ndimage.sobel(img_array)
                edge_density = np.mean(np.abs(edges))
                
                # Calculate texture variance
                texture_variance = np.var(img_array)
                
                # Combined complexity score
                complexity = edge_density * texture_variance
                complexity_scores.append(complexity)
                
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
    
    return complexity_scores

# Analyze background complexity across classes
plt.figure(figsize=(12, 5))

complexity_data = []
for class_name in CLASS_NAMES:
    complexities = analyze_background_complexity(class_name)
    complexity_data.append(complexities)

# Box plot comparison
plt.subplot(1, 2, 1)
plt.boxplot(complexity_data, tick_labels=CLASS_NAMES)
plt.ylabel('Background Complexity Score')
plt.title('Background Complexity by Class')

# Violin plot for distribution shape
plt.subplot(1, 2, 2)
violin_parts = plt.violinplot(complexity_data, showmedians=True)
plt.xticks([1, 2, 3], CLASS_NAMES)
plt.ylabel('Background Complexity Score')
plt.title('Complexity Distribution Shape')

# Color the violins
for pc, color in zip(violin_parts['bodies'], ['lightblue', 'lightcoral', 'lightgreen']):
    pc.set_facecolor(color)
    pc.set_alpha(0.7)

plt.tight_layout()
plt.show()

# Statistical summary
print("Background Complexity Summary:")
for i, class_name in enumerate(CLASS_NAMES):
    print(f"{class_name}: μ={np.mean(complexity_data[i]):.2f}, σ={np.std(complexity_data[i]):.2f}, "
          f"range=[{np.min(complexity_data[i]):.2f}, {np.max(complexity_data[i]):.2f}]")
```

Our background analysis reveals notable differences in environmental complexity across classes:
- Lacta: Highest complexity (μ=426,636) with wide variability, suggesting diverse, often cluttered environments
- Both: Moderate complexity (μ=391,591) with substantial range, indicating varied settings when both cats appear together
- Karamela: Lowest complexity (μ=358,675), often photographed in simpler, less cluttered backgrounds

The complexity gradient presents a potential learning shortcut—the model could associate cluttered backgrounds with Lacta and simpler environments with Karamela, rather than learning their distinctive visual features. The 19% complexity difference between Lacta and Karamela backgrounds provides a meaningful but superficial signal the model might exploit.

This finding informs our data augmentation strategy:
- Random cropping to force focus on cat regions rather than full backgrounds
- Background augmentation techniques if needed (though careful not to alter cat features)
- Combined with our previous augmentations, this ensures the model learns robust, background-invariant features

The substantial variability within each class is actually beneficial—it suggests enough background diversity to encourage the model to focus on the cats themselves rather than environmental cues.

# 9. Data Splits

With our data analysis complete, we now prepare the dataset for model development by creating train/validation/test splits. We use a stratified approach that maintains the original class distribution in each split, ensuring our models are evaluated on representative data.

The 70/15/15 split provides substantial training data while reserving meaningful portions for validation and final testing. This careful partitioning at the outset ensures all subsequent model experiments use identical data splits, enabling fair performance comparisons across different architectures and techniques.

```python
def create_stratified_splits(data_path, output_path, test_size=0.15, val_size=0.15):
    """
    Create train/val/test splits while maintaining class distribution
    """
    # First create train+val vs test split, then split train+val into train vs val
    for class_name in CLASS_NAMES:
        class_path = os.path.join(data_path, class_name)
        image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        
        # First split: train+val vs test
        train_val_files, test_files = train_test_split(
            image_files, test_size=test_size, random_state=42
        )
        
        # Second split: train vs val
        train_files, val_files = train_test_split(
            train_val_files, test_size=val_size/(1-test_size), random_state=42
        )
        
        # Create directories and copy files
        for split, files in [('train', train_files), ('val', val_files), ('test', test_files)]:
            split_class_path = os.path.join(output_path, split, class_name)
            os.makedirs(split_class_path, exist_ok=True)
            
            for file in files:
                src = os.path.join(class_path, file)
                dst = os.path.join(split_class_path, file)
                shutil.copy2(src, dst)
        
        print(f"{class_name}: Train={len(train_files)}, Val={len(val_files)}, Test={len(test_files)}")

# Create the splits
create_stratified_splits(DATA_PATH, "../data/processed")
```
